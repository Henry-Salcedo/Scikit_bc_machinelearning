```python
import sklearn
from sklearn import datasets   
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
# Testing how sklearn datasets work for this project. I will be using the breast cancer wisconsin dataset.
```

Creating Data Split (Train/Test) 

```python
b_cancer = datasets.load_breast_cancer()
X = b_cancer.data
y = b_cancer.target
```

```python
# Create a machine learning classification models that then uses a standard data division process.
x_train, x_test, y_train, y_test = model_selection.train_test_split(b_cancer.data, b_cancer.target, test_size=0.2, random_state=18)
```

Setting up the 3 models.

```python
#  Logistic regression, remember that it is not a linear regression but a classification model. It is used for binary classification problems, where the output is a probability that the input belongs to a certain class.
log_reg = LogisticRegression(random_state=43, solver='liblinear')
log_reg.fit(x_train, y_train)
```

```python
# K-nearest neighbors, it is a non-parametric method used for classification and regression. It is a type of instance-based learning, where the function is only approximated locally and all computation is deferred until function evaluation.
knn = sklearn.neighbors.KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
# Random forest, it is an ensemble learning method for classification and regression that operates by constructing a multitude of decision trees.
random_forest = RandomForestClassifier(n_estimators=200, random_state=28)
random_forest.fit(x_train, y_train) 
```

Predicting Scores

```python
# Now to test the models and see how they perform on the test sets.
reg_test_pred = log_reg.predict(x_test)
knn_test_pred = knn.predict(x_test)
rf_test_pred = random_forest.predict(x_test)
```

```python
# Seeing the accuracy of the models scores on the test set.

log_reg_accuracy = accuracy_score(y_test, reg_test_pred)
knn_accuracy = accuracy_score(y_test, knn_test_pred)
rf_accuracy = accuracy_score(y_test, rf_test_pred)

# Then the percision is tested:
log_reg_precision = sklearn.metrics.precision_score(y_test, reg_test_pred)
knn_precision = sklearn.metrics.precision_score(y_test, knn_test_pred)
rf_precision = sklearn.metrics.precision_score(y_test, rf_test_pred)

# Next recall is tested:
log_reg_recall = sklearn.metrics.recall_score(y_test, reg_test_pred)
knn_recall = sklearn.metrics.recall_score(y_test, knn_test_pred)
rf_recall = sklearn.metrics.recall_score(y_test, rf_test_pred)
```

Printed Results (Accuarcy, Percision, Recall, and F1 Scores)

```python
# I splited the printing as to not overwhelm readers when looking them over.

print("Logistic Regression Accuracy:", log_reg_accuracy)
print("K-Nearest Neighbors Accuracy:", knn_accuracy)
print("Random Forest Accuracy:", rf_accuracy)
```

```python
print("Logistic Regression Precision:", log_reg_precision)
print("K-Nearest Neighbors Precision:", knn_precision)
print("Random Forest Precision:", rf_precision)
```

```python
print("Logistic Regression Recall:", log_reg_recall)
print("K-Nearest Neighbors Recall:", knn_recall)
print("Random Forest Recall:", rf_recall)
```

```python
print ("F1 Score for Logistic Regression:", sklearn.metrics.f1_score(y_test, reg_test_pred))
print ("F1 Score for K-Nearest Neighbors:", sklearn.metrics.f1_score(y_test, knn_test_pred))
print ("F1 Score for Random Forest:", sklearn.metrics.f1_score(y_test, rf_test_pred))
```

Reviewing Results Performance

```python
# Determining the quality of the models.

# The random forest model most often performed the best out of the 3 models in at least 3 out of the 4 scoring metrics.
# Random forest additionally had the highest precision score, showing it had the lowest false positive rates.
# The recall scores also showed that K-nearest neighbors had the highest recall score, showing it had the lowest false negative rates.
# Random forest had the best performance, with the highest F1 score.
# Thus for theses 3 models, random forest out performers the other 2 models scores a majority of the time, meaning that its the best performing model with the limitations. (Read the README file in GitHub for more details).
```
